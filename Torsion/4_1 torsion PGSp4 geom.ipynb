```python
# Define the polynomial for omega
R = PolynomialRing(QQ, "t")
f = R.gen()^2 - R.gen() + 1
K = NumberField(f, "omega")
omega = K.gen()
rt3 = -(2 * omega -1 )

# Solution of Ptolemy relations [Zickert 2020]
f10 = f20 = f12 = 1
f23 = -1
f11 = 1/2 * (-1 + rt3)
f21 = 1/2 * (1 + rt3)
f13 = f22 = -f21
x = 1/3 * (1 + rt3)
y = 3/8 * (-1 + rt3)
z = -1/3 * (1 + rt3)
w = 3/4
f22 = 3/2 * z
```

```python
tet0 = Conf4(K, [z, x, w,  w, x, y,  y, y, -z,  -z, -x, w,  f10, f20, f11, f21, f12, f22, f13, f23])
tet1 = Conf4(K, [x, z, y,  y, -z, w,  w, w, x,  -x, z, y,  f12, f22, f13, f23, f10, f20, f11, f21])
G = tet0.T012.G
```

```python
builder = TorsionComplexBuilder(tet0, tet1, G)
basis_index = builder.indices.basis_index
cell0_index = builder.indices.cell0_index
cell1_index = builder.indices.cell1_index
cell2_index = builder.indices.cell2_index
cell3_index = builder.indices.cell3_index
d1_matrix, d2_matrix, d3_matrix = builder.differential_matrices()
```

```python
l = tet0.beta(2, 1, 3) * tet0.beta(3, 2, 1) * tet0.beta(1, 2, 0) * tet0.beta(0, 1, 2)
m = tet0.beta(0, 1, 3)
```

```python
fixed = matrix(G.base_ring, [[G.coefficient(m.Ad(G.lie_basis(i)) - G.lie_basis(i), j) for j in basis_index] + [G.coefficient(l.Ad(G.lie_basis(i)) - G.lie_basis(i), j) for j in basis_index] for i in basis_index])
```

```python
fixed_vector = []
for b in fixed.kernel().basis():
    x = reduce(lambda x, y:x+y, (b[i] * G.lie_basis(basis_index[i]) for i in range(len(basis_index))), GenSympLie.zero(G.base_ring, G.n, J=GenSympGroup.standard_J(G.base_ring, G.n)))
    fixed_vector.append(x)
```

```python
fixed_vector
```

```python
h10 = builder._to_vector(builder._d_aux(fixed_vector[0], [G.identity()], [('short', 1)], [-1]), cell1_index)
h11 = builder._to_vector(builder._d_aux(fixed_vector[1], [G.identity()], [('short', 1)], [-1]), cell1_index)
h20 = builder._to_vector(builder._d_aux(fixed_vector[0], [G.identity()]*8, [('triangle', k) for k in range(8)], [1]*4 + [-1]*4), cell2_index)
h21 = builder._to_vector(builder._d_aux(fixed_vector[1], [G.identity()]*8, [('triangle', k) for k in range(8)], [1]*4 + [-1]*4), cell2_index)
```

```python
d2_basis = block_matrix(2, 1, [d2_matrix, identity_matrix(len(cell2_index))]).column_space().basis()
bd2_basis0 = [v[len(cell1_index):] for v in d2_basis][:d2_matrix.rank()]
bd2_basis = [v[:len(cell1_index)] for v in d2_basis][:d2_matrix.rank()]
d1_basis = block_matrix(2, 1, [d1_matrix, identity_matrix(len(cell1_index))]).column_space().basis()
bd1_basis0 = [v[len(cell0_index):] for v in d1_basis][:d1_matrix.rank()]
bd1_basis = [v[:len(cell0_index)] for v in d1_basis][:d1_matrix.rank()]
```

```python
d3_basis = block_matrix(2, 1, [d3_matrix, identity_matrix(len(cell3_index))]).column_space().basis()
bd3_basis0 = [v[len(cell2_index):] for v in d3_basis][:d3_matrix.rank()]
bd3_basis = [v[:len(cell2_index)] for v in d3_basis][:d3_matrix.rank()]
d0_basis = block_matrix(2, 1, [d0_matrix, identity_matrix(len(cell0_index))]).column_space().basis()
bd0_basis0 = [v[0:] for v in d0_basis][:d0_matrix.rank()]
bd0_basis = [v[:0] for v in d0_basis][:d0_matrix.rank()]
```

```python
det0 = matrix(bd1_basis)
det1 = matrix(bd2_basis + [h10, h11] + bd1_basis0)
det2 = matrix(bd3_basis + [h20, h21] + bd2_basis0)
det3 = matrix(bd3_basis0)
```

```python
torsion = 1/det0.det(algorithm='df') * det1.det(algorithm='df') / det2.det(algorithm='df') * det3.det(algorithm='df')
```
